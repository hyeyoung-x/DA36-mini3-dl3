```python
import os
import shutil

def unpack_all_files(src_dir, dest_dir):

    # 대상 디렉토리가 없으면 생성
    os.makedirs(dest_dir, exist_ok=True)
    
    # 하위 디렉토리 탐색
    for root, dirs, files in os.walk(src_dir):
        for file in files:
            src_file_path = os.path.join(root, file)
            dest_file_path = os.path.join(dest_dir, file)
            
            # 파일 복사
            shutil.copy(src_file_path, dest_file_path)
            print(f"Copied: {src_file_path} -> {dest_file_path}")

# 사용 예시
source_directory = "C:\\Users\\playdata2\\Desktop\\소파40"  # 원본 디렉토리 경로 작성
destination_directory = "C:\\Users\\playdata2\\Desktop\\소파40s"  # 대상 디렉토리 경로 작성

unpack_all_files(source_directory, destination_directory)
```
